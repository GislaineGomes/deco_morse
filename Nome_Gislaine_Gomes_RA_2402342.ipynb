```python
# Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# Passo 1: Gerar os dados
X = np.linspace(0.2, 8, 170)
Y1 = 5.6 + np.cos(X)
Y2 = 2.8 + np.cos(2.8 + X / 0.5) / 2
Y3 = np.random.uniform(Y1, Y2, len(X))

# Passo 2: Criar o grfico base
plt.figure(figsize=(10, 6))

# Passo 3: Plotar as linhas com cores diferentes
plt.plot(X, Y1, label="Limite Superior+ 2402342", color='blue', linestyle='-', marker='x')
plt.plot(X, Y2, label="Limite Inferior+ 2402342", color='red', linestyle='--')
plt.plot(X, Y3, label="Aleatria", color='green', linestyle=':', marker='s')

# Passo 4: Ajustar os eixos e adicionar rtulos e ttulo
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.xlabel("X", fontsize=12)
plt.ylabel("Y", fontsize=12)
plt.title("Grfico - Nome: Gislaine Gomes  - RA: 2402342", fontsize=14, fontweight='bold')

# Passo 5: Adicionar divises menores nos eixos
plt.xticks(np.arange(0, 6.2, 0.2), fontsize=10)
plt.yticks(np.arange(0, 6.2, 0.2), fontsize=10)
plt.grid(which='both', linestyle='--', linewidth=0.5, alpha=0.7)

# Passo 6: Adicionar um crculo no ponto (2.0, 2.0) com a etiqueta do RA
plt.scatter(2.0, 2.0, color='purple', label='Ponto (2.0, 2.0)', edgecolor='black', s=100)
plt.text(2.0, 1.7, '2402342', color='black', fontsize=10, ha='center', va='center', fontweight='bold')

# Passo 7: Adicionar a legenda
plt.legend(fontsize=10, loc='lower center', frameon=True, shadow=True, edgecolor='gray', bbox_to_anchor=(0.5, -0.15))

# Passo 8: Ajustar o layout e exibir o grfico
plt.tight_layout()
plt.savefig("Grafico_Funcoes.png", bbox_inches='tight')
plt.show()
```
